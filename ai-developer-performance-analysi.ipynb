```python
from sklearn.model_selection import train_test_split
```

```python
X = AI.drop(columns = ["Cognitive_Load", "Bugs_Found", "Bugs_Fixed","Lines_of_Code"])
y = AI["Hours_Coding"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 12, test_size = 0.2)
```

```python
X_train.describe().round(2)
```

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
Scaler.fit(X_train)

X_train_scaled = Scaler.transform(X_train)
X_test_scaled = Scaler.transform(X_test)

X_train_Scaler = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_Scaler = pd.DataFrame(X_test_scaled, columns = X_train.columns)

X_test_Scaler.describe()
```

        <p style="margin:0 0 10px;">
          🚀 <strong>SimpleImputer</strong> is a tool in Python from <strong>scikit-learn</strong> used to fill
          (impute) missing values in a dataset.
        </p>

<p style="margin:0 0 10px;">
          If your dataset does not contain missing values, SimpleImputer will not have any effect
          because there is nothing to fill.
        </p>

<ul style="margin:6px 0 0 18px;">
          <li><strong>mean</strong> → replaces missing values with the column mean</li>
          <li><strong>median</strong> → replaces missing values with the median</li>
          <li><strong>most_frequent</strong> → replaces missing values with the most frequent value</li>
          <li><strong>constant</strong> → replaces missing values with a user-defined constant</li>
        </ul>
